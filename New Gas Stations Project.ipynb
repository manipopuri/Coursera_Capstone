```python
import pandas as pd
read_gas_stations = pd.read_csv("County_Fueling_Stations.csv")
read_gas_stations.head()
gas_stations = read_gas_stations[["OBJECTID","city",'addrln1','zip','latitude','longitude']]
gas_stations
```

```python
nigh = pd.read_csv('zip_codes_states.csv')
nigh.head()
nigh
```

```python
result = pd.merge(gas_stations, nigh, how='right', left_on='zip', right_on='zip_code')
result
new_df = result[['OBJECTID','city_y','addrln1','zip','latitude_y','longitude_y','state','county']]
new_df
```

```python
from bs4 import BeautifulSoup as bsoup
import requests
import lxml.html as lh

response_obj = requests.get('https://en.wikipedia.org/wiki/Demographics_of_Los_Angeles_County#Housing').text
soup = bsoup(response_obj,'lxml')
la_county = soup.find('table',{'class':'wikitable collapsible collapsed sortable'})

totals=la_county.find_all('tr')
nrows=len(totals)
nrows

header=totals[0].text.split()
header

totals[5].text

totals[5].text.split('\n')

records =[]
n=1
while n < nrows :
    city=totals[n].text.split('\n')[1]
    Population=totals[n].text.split('\n')[3]
    #Population=totals[n].text.split('\n')[7]
    records.append((city,Population))
    n=n+1

pop=pd.DataFrame(records, columns=['city', 'Population'])
pop.head(1000)
```

```python
new_set = pd.merge(new_df, pop, how='left', left_on='city_y', right_on='city')
new_set
new_df1 = new_set[['OBJECTID','city_y','addrln1','zip','latitude_y','longitude_y','state','county','Population']]
new_df1
```

```python
new_df1['zip'].value_counts()
nigh2 = new_df1[(new_df1.OBJECTID.notnull())]
nigh2 = nigh2[(nigh2.Population.notnull())]
nigh2 = nigh2[(nigh2.latitude_y.notnull())]
nigh2
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
%matplotlib inline
import seaborn as sns

nigh2 = nigh2[(nigh2['city_y'] != 'Los Angeles')] 
nigh2.groupby('city_y')['OBJECTID'].count().plot.bar(figsize=(20,5), color = 'Blue')
plt.title('Gas Stations by Borough : LA', fontsize = 20)
plt.xlabel('Borough', fontsize = 15)
plt.ylabel('No.of Gas Stations', fontsize=15)
plt.xticks(rotation = 'vertical')
plt.show()

```

```python
population = nigh2[['city_y','Population']]

population = population.drop_duplicates()


a = nigh2[['Population']]
population.info()
population['Population'] = population['Population'].str.replace(',', '')
population['Population'] = population['Population'].astype(int)

population.info()
population = population[(population['city_y'] != 'Los Angeles')]


population.groupby('city_y').sum().plot.bar(figsize=(20,5), color = 'Blue')
plt.title('Population by Borough : LA', fontsize = 20)
plt.xlabel('Borough', fontsize = 15)
plt.ylabel('population', fontsize=15)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
gas_city =  nigh2[['city_y','OBJECTID']]
gas_city = gas_city.groupby('city_y').count()
gas_city

gas_city_pop = pd.merge(gas_city, population, how='left', left_on='city_y', right_on='city_y')
gas_city_pop



gas_city_pop['avg'] = gas_city_pop.apply(lambda row: row.Population / row.OBJECTID, axis=1)
gas_city_pop.sort_values(by=['avg'])

gas_city_pop = pd.merge(gas_city_pop, nigh, how='left', left_on='city_y', right_on='city')
gas_city_pop = gas_city_pop[['city','OBJECTID','Population','latitude','longitude','state','county','avg']]


gas_city_pop_sort = gas_city_pop.sort_values(by=['avg'])
gas_city_pop_sort.drop_duplicates()
gas_city_slect = gas_city_pop_sort.tail()
```
