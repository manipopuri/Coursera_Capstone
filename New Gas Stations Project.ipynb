```python
import pandas as pd

read_gas_stations = pd.read_csv("County_Fueling_Stations.csv")
read_gas_stations.head()
gas_stations = read_gas_stations[["OBJECTID","city",'addrln1','zip','latitude','longitude']]
gas_stations
```

```python
nigh = pd.read_csv('zip_codes_states.csv')
nigh.head()
nigh
```

```python
result = pd.merge(gas_stations, nigh, how='right', left_on='zip', right_on='zip_code')
result
new_df = result[['OBJECTID','city_y','addrln1','zip','latitude_y','longitude_y','state','county']]
new_df
```

```python
from bs4 import BeautifulSoup as bsoup
import requests
import lxml.html as lh

response_obj = requests.get('https://en.wikipedia.org/wiki/Demographics_of_Los_Angeles_County#Housing').text
soup = bsoup(response_obj,'lxml')
la_county = soup.find('table',{'class':'wikitable collapsible collapsed sortable'})

totals=la_county.find_all('tr')
nrows=len(totals)
nrows

header=totals[0].text.split()
header

totals[5].text

totals[5].text.split('\n')

records =[]
n=1
while n < nrows :
    city=totals[n].text.split('\n')[1]
    Population=totals[n].text.split('\n')[3]
    #Population=totals[n].text.split('\n')[7]
    records.append((city,Population))
    n=n+1

pop=pd.DataFrame(records, columns=['city', 'Population'])
pop.head(1000)
```

```python
new_set = pd.merge(new_df, pop, how='left', left_on='city_y', right_on='city')
new_set
new_df1 = new_set[['OBJECTID','city_y','addrln1','zip','latitude_y','longitude_y','state','county','Population']]
new_df1
```

```python
new_df1['zip'].value_counts()
nigh2 = new_df1[(new_df1.OBJECTID.notnull())]
nigh2 = nigh2[(nigh2.Population.notnull())]
nigh2 = nigh2[(nigh2.latitude_y.notnull())]
nigh2

nigh2['city_y'].value_counts()
df1 = nigh2[['city_y','latitude_y','longitude_y']]
df2 = df1.drop_duplicates()
df2 = df2[(df2['city_y'] != 'Los Angeles')]
df2
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
import json # library to handle JSON files
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

venues = results['response']['groups'][0]['items']  
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues
```
